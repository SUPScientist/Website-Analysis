### Look at EPA stations first

# **NB: Datashader doesn't represent these sparse data all that well, at least not how I used it...**

```python
import pandas as pd
import numpy as np
import os

import datashader as ds
import datashader.transfer_functions as tf
from datashader.utils import lnglat_to_meters as webm

import cmocean
import colorcet
```

```python
filename = 'annual_conc_by_monitor_2019.csv'
filepath = os.path.join('data', filename)
df_epa = pd.read_csv(filepath)
df_epa.describe()
```

```python
# How many unique sites are there?
len(df_epa.loc[:, 'Site Num'].unique())
```

### We'll lose non-numeric data but this will give us lat/lon for unique `Site Num`

```python
df_epa_stations = df_epa.groupby(by = 'Site Num').mean()
df_epa_stations.head()
```

### Make some datashader plots
- Convert lat/lon to Web Mercator
- Make canvas and plot

```python
df_epa_stations.loc[:, 'easting'], df_epa_stations.loc[:, 'northing'] = webm(df_epa_stations['Longitude'], df_epa_stations['Latitude'])
df_epa_stations.head()
```

```python
df_epa_stations.northing.isna().sum()
```

```python
cmap = cmocean.cm.thermal

cvs = ds.Canvas(plot_width=400, plot_height=400)
agg = cvs.points(df_epa_stations, 'Longitude', 'Latitude')
tf.set_background(tf.shade(agg, cmap=colorcet.fire),"black")

```
